```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d, Axes3D
```

```python
results = np.loadtxt('asian.txt')

M1 = 50
M2 = 50

x0 = 0.
x1 = 1.

xi, b = np.meshgrid(np.linspace(x0, x1, M2+1), np.linspace(0., 1., M1+1))
U = np.reshape(results, (M1+1, M2+1))

fig = plt.figure()
ax = Axes3D(fig)
ax.plot_wireframe(xi, b, U)
ax.set_xlabel('$b$')
ax.set_ylabel('$\\xi$')
ax.set_zlabel('$U(t, \\xi, b)$')
ax.view_init(elev=None, azim=225)
```
